```python
import csv
import urllib.request

url_csv = "https://raw.githubusercontent.com/fiesc-junior-prado/mine_projeto_bloco_1/refs/heads/main/olist_products_dataset.csv"

dados_extraidos = []
cabecalho = []

resposta = urllib.request.urlopen(url_csv)

linhas_texto = [linha.decode('utf-8') for linha in resposta.readlines()]

leitor = csv.DictReader(linhas_texto)

dados_extraidos = list(leitor)

cabecalho = leitor.fieldnames

print(len(dados_extraidos))

print(dados_extraidos[0])

## Exportar CSV

nome_arquivo_saida = "export_copia_produtos_olist.csv"

with open(nome_arquivo_saida, mode='w', encoding='utf-8', newline='') as arquivo_local:

    escritor = csv.DictWriter(arquivo_local, fieldnames=cabecalho)

    escritor.writeheader()

    escritor.writerows(dados_extraidos)
```

```python
nome_arquivo_saida = "arquivo_saida.csv"

def salvar_csv(dados_extraidos, nome_arquivo_saida):
    if not dados_extraidos:
        print("Não há dados para salvar.")
        return
    colunas = dados_extraidos[0].keys()
    with open(nome_arquivo_saida, mode="w", newline="", encoding="utf-8") as arquivo:
        writer = csv.DictWriter(
            arquivo,
            fieldnames=colunas
        )
        writer.writeheader()
        writer.writerows(dados_extraidos)

salvar_csv(dados_extraidos, nome_arquivo_saida)
```
