```python
import time
import random

print("Gerando 1 MILHÃO de acessos de usuários no servidor...")
# Gera 1 milhão de acessos, mas os IDs dos clientes vão só do 1 ao 500 (muita repetição)
acessos_servidor = [random.randint(1, 500) for _ in range(1000000)]

print(f"Total de cliques registrados: {len(acessos_servidor)}\n")
print("Filtrando usuários únicos...")

# Marcando o tempo e usando o SET
inicio = time.time()

usuarios_unicos = set(acessos_servidor)

fim = time.time()
tempo_gasto = fim - inicio

# Resultado
print(f"Usuários reais (únicos) encontrados: {len(usuarios_unicos)}")
print(f"Tempo gasto para varrer 1 MILHÃO de registros: {tempo_gasto:.5f} segundos")
```

```python
# Criando uma base de dados vazia (Lista)
tabela_clientes = []

# Preenchendo com 50.000 registros (Dicionários)
print("Baixando dados do servidor...")
for id_cliente in range(50000):
    registro = {
        "id": id_cliente,
        "nome": f"Cliente VIP {id_cliente}",
        "score_credito": random.randint(300, 1000),
        "ativo": True
    }
    tabela_clientes.append(registro)

print(f"Base carregada com sucesso! Total de linhas: {len(tabela_clientes)}\n")

# Consultando um dado específico no meio do Big Data
linha_desejada = 45999
print("--- DADOS DO CLIENTE 45999 ---")
print(f"Nome: {tabela_clientes[linha_desejada]['nome']}")
print(f"Score: {tabela_clientes[linha_desejada]['score_credito']}")
```

```python
import csv
import urllib.request

# 1. Puxando o CSV do Titanic direto do GitHub (Dados Brutos)
print("Baixando dataset do Titanic da Internet...")
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
resposta = urllib.request.urlopen(url)
linhas = [linha.decode('utf-8') for linha in resposta.readlines()]

# 2. Transformando em uma Lista de Dicionários!
leitor = csv.DictReader(linhas)
passageiros = list(leitor)

print(f"Sucesso! Temos os dados de {len(passageiros)} passageiros.\n")

# 3. Analisando o Tipo de Dado do primeiro passageiro
print("--- DADO BRUTO ---")
print(f"Nome: {passageiros[0]['Name']}")
print(f"Idade do passageiro 0: {passageiros[0]['Age']}")
print(f"TIPO PRIMITIVO DA IDADE: {type(passageiros[0]['Age'])}\n") # Vai mostrar STRING!

# 4. Limpando os dados com o Try/Except e List Comprehension
print("--- LIMPEZA E CONVERSÃO (CASTING) ---")
idades_limpas = []

for p in passageiros:
    try:
        # Tenta converter a string para Float
        idade_numero = float(p['Age'])
        idades_limpas.append(idade_numero)
    except ValueError:
        # Alguns passageiros não têm a idade preenchida (está vazio).
        # O try/except evita que o Colab exploda aqui!
        continue

print(f"Conseguimos converter as idades de {len(idades_limpas)} passageiros.")
print(f"Exemplo das 5 primeiras idades (agora como Float): {idades_limpas[:5]}")
```
